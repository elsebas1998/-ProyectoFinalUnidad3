```python
#Importacion de las librerias
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy import MetaData
from datetime import datetime
import psycopg2
import pyodbc
```

```python
engine = create_engine('sqlite:///Northwind.sqlite')
```

```python
dfcustomer = pd.read_sql_query("""SELECT Id as Id_Customer,CompanyName, ContactName, ContactTitle, Address, City,Region,PostalCode,Country,Phone,Fax FROM Customer; """, con=engine.connect())
dfcustomer
```

```python
dflocation = pd.read_sql_query("""SELECT ShipVia,ShipName,ShipAddress,ShipCity,ShipRegion,ShipPostalCode,ShipCountry FROM Orders; """, con=engine.connect())
dflocation 
```

```python
dfemployees = pd.read_sql_query("""SELECT LastName,FirstName,Title,TitleOfCourtesy,BirthDate,HireDate,Address,City,Region,PostalCode,Country,Extension,Notes,PhotoPath,t.TerritoryDescription,r.RegionDescription FROM Employee e
    INNER JOIN
    EmployeeTerritory et  
    INNER JOIN
    Territory t ON t.Id = et.TerritoryId
    INNER JOIN
    Region r ON r.Id = t.RegionId    
    ; """, con=engine.connect())
dfemployees
```

```python
dfproducts = pd.read_sql_query("""SELECT ProductName,CategoryName,Description,QuantityPerUnit as QuanntityPerUnit,UnitPrice,UnitsInStock,UnitsOnOrder,ReorderLevel,Discontinued as Discount FROM Product p
    INNER JOIN
    Category c ON c.Id = p.CategoryId
    ; """, con=engine.connect())
dfproducts
```

```python
dfsuppliers = pd.read_sql_query("""SELECT CompanyName,ContactName,ContactTitle,Address,City,Region,PostalCode,Country,Phone,Fax FROM Customer; """, con=engine.connect())
dfsuppliers
```

```python
dfDim = pd.read_sql_query("""SELECT OrderDetail.UnitPrice,OrderDetail.Quantity,OrderDetail.Discount,Orders.Freight,Customer.Id as IdCustomer,Employee.Id as IdEmployee,Orders.Id as IdLocation,Supplier.Id as IdSupplier,Product.Id as IdProduct,IdTime as IdTime 
    FROM Orders
    INNER JOIN
    OrderDetail ON OrderDetail.OrderId = Orders.Id
    INNER JOIN
    Customer ON Customer.Id = Orders.CustomerId
    INNER JOIN
    Employee ON Employee.Id = Orders.EmployeeId
    INNER JOIN
    Product ON Product.Id = OrderDetail.ProductId
    INNER JOIN
    Supplier ON Supplier.Id = Product.SupplierId 
    INNER JOIN 
    Dim_Time ON Orders.OrderDate= substr(Dim_Time.date,0,11)
    ; """, con=engine.connect())
dfDim
```
